```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('spotify-2023.csv', delimiter=',')
```

Сортуємо пісні 

```python
sort = data.sort_values(by='streams', ascending=False)
```

```python
sort_data = sort.head(10)
```

 Побудувати горизонтальну стовпчикову діаграму, на якій відображаються 10 
найпопулярніших пісень на Spotify у 2023 році за кількістю 
прослуховувань

```python
plt.figure(figsize=(15, 8))
plt.barh(sort_data['track_name'], sort_data['streams'], color='skyblue')
plt.xlabel('Кількість прослуховувань')
plt.ylabel('Назва пісні')
plt.title('Топ 10 найпопулярніших пісень')
plt.show()
```

Побудувати кругову діаграму для тих самих даних

```python
plt.figure(figsize=(10, 10))
plt.pie(sort_data['streams'], labels=sort_data['track_name'], autopct='%1.1f%%')
plt.title('Топ 10 найпопулярніших пісень на Spotify у 2023 році')
plt.show()
```

 Створіть кругову діаграму або секторну діаграму, яка відображає розподіл
за danceability серед найпопулярніших пісень у 2023 році.

```python
plt.figure(figsize=(10, 10))
plt.pie(sort_data['danceability_%'], labels=sort_data['track_name'], autopct='%1.1f%%')
plt.title('Розподіл за показником danceability серед найпопулярніших пісень')
plt.show()
```

 Створіть стовпчикову діаграму, на якій відображаються 10 найпопулярніших 
артистів на Spotify у 2023 році за кількістю слухачів.

```python
plt.figure(figsize=(15, 6))
plt.bar(sort_data['artist(s)_name'], sort_data['streams'], color='skyblue')
plt.xlabel('Артист')
plt.ylabel('Кількість слухачів')
plt.title('10 найпопулярніших артистів')
plt.tight_layout()
plt.show()
```
